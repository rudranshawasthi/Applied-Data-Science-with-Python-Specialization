```python
df1 = pd.read_csv('covid_19_india.csv')
df1 = df1.drop(['ConfirmedIndianNational', 'ConfirmedForeignNational','Time','Sno'], axis = 1)
df1 = df1.set_index(['Date','State/UnionTerritory'])
# df1 = df1.set_index(['location','date'])
# df1.loc['India']
```

```python
df1 = df1.drop('Cases being reassigned to states',level = 'State/UnionTerritory')
df1['Active'] = df1['Confirmed']-(df1['Cured']+df1['Deaths'])
```

```python
#df1 = df1.drop(['Cured','Deaths'],axis =1)
df1 = df1.reset_index()
```

```python
State = df1[df1['State/UnionTerritory']=='Kerala']
State = State.drop('State/UnionTerritory',axis=1)
date = State['Date']
mi = State[70:].groupby('Active').agg(min).iloc[0]['Date']
State = State.drop('Date',axis = 1)
# mi = mi['Date']
mi
```

```python
%matplotlib notebook
plt.figure()
plt.style.use('seaborn')
_ = plt.plot(date,State['Confirmed'],label = 'Confirmed',alpha = 0.5)
_ = plt.bar(date,State['Active'], label = 'Active',alpha=.5)
_ = plt.plot(date,State['Cured'],label='Cured',alpha= .8)
_ = plt.xticks(date[::20])
# plt.axvline(x=mi,color = 'grey')
ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.tick_params(axis='both', which='both', length=0)
plt.legend(frameon=False)
plt.title('COVID19 status in Kerala')
plt.xlabel('Dates')
plt.ylabel('People')
```
